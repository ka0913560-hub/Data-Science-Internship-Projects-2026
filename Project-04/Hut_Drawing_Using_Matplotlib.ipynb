```python
#Make a hut using MATPLOTLIB

import matplotlib.pyplot as plt
from matplotlib.patches import Circle

x1=[2,2,4,4,2]
y1=[2,7,7,2,2]
plt.plot(x1,y1)

x2=[4,4,8,8,4]
y2=[2,7,7,2,2]
plt.plot(x2,y2)

x3=[2,3,4,2]
y3=[7,10,7,7]
plt.plot(x3,y3)

x4=[3,8,8,4,3]
y4=[10,10,7,7,10]
plt.plot(x4,y4)

x5=[2.5,2.5,3.5,3.5]
y5=[3,6,6,3]
plt.plot(x5,y5)


# Circular window on roof
circle = Circle(
    (3, 8),
    0.45,
    facecolor="yellow",
    edgecolor="black",
    linewidth=2,
    zorder=10
)
plt.gca().add_patch(circle)

plt.savefig("hut.jpeg")

# House body color
plt.fill(x1, y1, color="bisque")

# Second room color
plt.fill(x2, y2, color="wheat")

# Roof colors
plt.fill(x3, y3, color="firebrick")
plt.fill(x4, y4, color="indianred")
# Door color
plt.fill(x5, y5, color="saddlebrown")

# Window
plt.fill([5,6,6,5],[4,4,5,5], color="skyblue")
plt.plot([5,6,6,5,5],[4,4,5,5,4], color="black")
plt.plot([5.5,5.5],[4,5], color="black")
plt.plot([5,6],[4.5,4.5], color="black")



# Remove axis numbers
plt.xticks([])
plt.yticks([])

# Equal aspect ratio
plt.axis("equal")

# Title
plt.title("🏠 Colorful Hut using Matplotlib",fontsize=16,fontweight="bold")

plt.savefig("hutt.jpeg")
plt.show()
```
