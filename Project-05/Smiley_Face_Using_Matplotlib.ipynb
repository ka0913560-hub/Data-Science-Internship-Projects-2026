```python
#Smiley_Face_Using_Matplotlib

import matplotlib.pyplot as plt
from matplotlib.patches import Circle
from matplotlib.patches import Circle, Arc

circle=Circle([.5,.6],0.38,facecolor="gold",edgecolor="black",linewidth=3)
plt.gca().add_patch(circle)

circle=Circle([.33,.75],0.05,color="black")
plt.gca().add_patch(circle)

circle=Circle([.66,.75],0.05,color="black")
plt.gca().add_patch(circle)

smile = Arc((0.5,0.42),width=0.25,height=0.15,angle=0,theta1=200,theta2=340,color="black",linewidth=3)
plt.gca().add_patch(smile)

plt.title("😊 Smiley Face using Matplotlib",fontsize=16,fontweight="bold")
plt.savefig("Smiley.jpeg")
plt.show()
```
