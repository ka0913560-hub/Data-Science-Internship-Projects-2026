```python
# Project on numpy-
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
img=Image.open(r"s4.jpeg")
img_arr=np.array(img)
img_arr[250:,350:,1]=0    #fourth
img_arr[0:250,0:350,0]=0 #first
img_arr[0:250,350:,2]=0  #second
img_arr[250:,0:350,0]=250#third

plt.imshow(img_arr)
```
