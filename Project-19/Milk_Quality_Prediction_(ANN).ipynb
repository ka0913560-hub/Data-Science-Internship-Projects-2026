Milk Quality Prediction using Artificial Neural Network (ANN)

```python
import numpy as np
import pandas as pd
import tensorflow as tf

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler

from keras.models import Sequential
from keras.layers import Dense
```

```python
dataset = pd.read_csv("milknew.csv")

dataset.head()
```

```python
dataset.info()
```

```python
dataset.describe()
```

```python
dataset.isnull().sum()
```

```python
le = LabelEncoder()

dataset["Grade"] = le.fit_transform(dataset["Grade"])
```

```python
dataset.head()
```

```python
X = dataset.drop("Grade", axis=1)

y = dataset["Grade"]
```

```python
train_X, test_X, train_y, test_y = train_test_split(
    X,
    y,
    test_size=0.3,
    random_state=123
)
```

```python
sc = StandardScaler()

train_X = sc.fit_transform(train_X)

test_X = sc.transform(test_X)
```

```python
#Build ANN
model = Sequential()
model.add(Dense(units=8,activation="relu",input_dim=7))
model.add(Dense(units=6,activation="relu"))
model.add(Dense(units=4,activation="relu"))
model.add(Dense(units=3,activation="softmax"))
```

```python
#compile
model.compile(optimizer="adam",loss="sparse_categorical_crossentropy",metrics=["accuracy"])
```

```python
#train
model.fit(train_X,train_y,epochs=50,batch_size=32,validation_split=0.1)
```

```python
model.evaluate(test_X,test_y)
```

```python
predictions = model.predict(test_X)

predictions
```

```python
predicted_class = np.argmax(predictions, axis=1)

predicted_class
```

```python
model.save("milk_quality_ann.keras")
```
